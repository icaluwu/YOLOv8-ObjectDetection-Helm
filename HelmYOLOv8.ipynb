Step 4 : Ready YOLOv8 Nano for startup

````python
from ultralytics import YOLO

# Load model YOLOv8
model = YOLO('yolov8n.pt')  # Gunakan versi YOLOv8 (nano, small, medium, etc.)
````

Step 5 : Train model *data.yaml* with starter of YOLOv8


````python
# Melatih model
model.train(data='/content/dataset/data.yaml', epochs=100, imgsz=640, batch=16)
````

Step 7: Check Analytic Log for Analytic requirements

````python
import pandas as pd

# Lokasi file log pelatihan
log_path = "runs/detect/train/results.csv"
data = pd.read_csv(log_path)

# Tampilkan beberapa baris pertama data
print(data.head())
````

Step 8: Test *model.pt* from our drive using our ready-for-test images (CC License Images)

````python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Muat model yang sudah dilatih
model = YOLO('/content/dataset/model.pt')  # Gantilah dengan path model yang sesuai

# Gambar yang ingin diuji
image_path = '/content/drive/MyDrive/Samples/camera2.jpg'  # Gantilah dengan path gambar baru

# Melakukan prediksi pada gambar
results = model(image_path)

# Akses hasil prediksi (misalnya pertama dalam list)
result = results[0]

# Menampilkan gambar dengan bounding box
result.show()

# Jika ingin menampilkan gambar menggunakan matplotlib
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Mengubah format warna ke RGB untuk matplotlib
plt.imshow(img)
plt.axis('off')  # Menyembunyikan sumbu
plt.show()
````

Step 9: Make a Analytic Graphic for project successful testing requirements, and to make sure we have the accurate *epochs*

````python
import matplotlib.pyplot as plt

# Memplot training loss
plt.figure(figsize=(10, 6))
plt.plot(data['epoch'], data['train/box_loss'], label='Box Loss', color='blue')
plt.plot(data['epoch'], data['train/cls_loss'], label='Class Loss', color='orange')
plt.plot(data['epoch'], data['train/dfl_loss'], label='DFL Loss', color='green')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss Over Epochs')
plt.legend()
plt.grid()
plt.show()
````

````python
plt.figure(figsize=(10, 5))
plt.plot(data['epoch'], data['metrics/precision(B)'], label='Precision', color='blue')
plt.plot(data['epoch'], data['metrics/recall(B)'], label='Recall', color='orange')
plt.plot(data['epoch'], data['metrics/mAP50(B)'], label='mAP@50', color='green')
plt.plot(data['epoch'], data['metrics/mAP50-95(B)'], label='mAP@50-95', color='red')
plt.xlabel('Epoch')
plt.ylabel('Metrics')
plt.title('Precision, Recall, and mAP Over Epochs')
plt.legend()
plt.grid()
plt.show()
````

````python
plt.figure(figsize=(10, 5))
plt.plot(data['epoch'], data['val/box_loss'], label='Validation Box Loss', color='blue')
plt.plot(data['epoch'], data['val/cls_loss'], label='Validation Class Loss', color='orange')
plt.plot(data['epoch'], data['val/dfl_loss'], label='Validation DFL Loss', color='green')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Validation Loss Over Epochs')
plt.legend()
plt.grid()
plt.show()
````

````python
plt.figure(figsize=(10, 5))
plt.plot(data['epoch'], data['train/box_loss'], label='Train Box Loss', linestyle='--', color='blue')
plt.plot(data['epoch'], data['val/box_loss'], label='Val Box Loss', linestyle='-', color='blue')
plt.plot(data['epoch'], data['train/cls_loss'], label='Train Class Loss', linestyle='--', color='orange')
plt.plot(data['epoch'], data['val/cls_loss'], label='Val Class Loss', linestyle='-', color='orange')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train and Validation Loss Over Epochs')
plt.legend()
plt.grid()
plt.show()
````
